# 线性回归

回归分析是一种预测性的建模技术，它研究的是因变量（目标）和自变量（预测器）之间的关系。这种技术通常用于预测分析。线性回归是最简单的模型，模型假定因变量与自变量之间有着线性关系。

## 简单的线性回归
首先取一个最简单线性的回归，一元线性回归$y=a+bx$的例子，式子中，只有一个自变量x，因变量为y，a为截距，b为x所对应的系数。线性回归中我们需要在给定数据集$(x_1,y_1),(x_2, y_2),···,(x_n,y_n)$下，求解$a$, $b$的值，使得求解$a',b'$，能够更好的符合真正的线性关系。

下面我们创建一个带噪声的数据集，实现线性回归

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# 创建带噪声的数据
X = np.random.randn(50)
a_true = 1
b_true = 1
y = a_true + b_true * X + 0.1 * np.random.normal(0, 1, 50)
plt.scatter(X, y)
plt.show()
```

```python
# 线性回归, 
X = X.reshape((50, 1))
linear_regression = LinearRegression()
linear_regression.fit(X, y)
y_pred = linear_regression.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, 'r-', linewidth=2)
plt.show()
```

## 多元线性回归
当因变量为多个自变量的线性组合的时候，此时就是多元线性回归。此时模型为:$$y=\beta_0 + \beta_1x_1+\beta_2x_2+···+\beta_nx_n$$
简记为：$$y=\beta_0 + \sum_{i=1}^{n}{\beta_ix_i}$$
这里$\beta_i$是未知的参数，变量$x_i$是已知的数据，不管$x_i$的数据怎样，模型对于$\beta$来说，肯定是线性的。
